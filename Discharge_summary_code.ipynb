```python
#to get text files
import os
import pytesseract
from PIL import Image


# 1. TESSERACT PATH


pytesseract.pytesseract.tesseract_cmd = (
    r"C:\Users\Asus\Downloads\tesseract.exe"
)


# 2. IMAGE FOLDER PATH


folder_path = r"C:\Users\Asus\Documents\project_files\patient_images"


# 3. OUTPUT FILES


ocr_output_file = r"C:\Users\Asus\Documents\project_files\patient_text.txt"

summary_output_file = r"C:\Users\Asus\Documents\project_files\discharge_summary.txt"


# 4. OCR PROCESS


all_text = ""

files = sorted(os.listdir(folder_path))

for file in files:

    if file.lower().endswith(".png"):

        image_path = os.path.join(folder_path, file)

        print("Reading:", file)

        try:

            image = Image.open(image_path)

            text = pytesseract.image_to_string(image)

            all_text += "\n\n"
            all_text += "=" * 50
            all_text += "\n"
            all_text += f"PAGE: {file}"
            all_text += "\n"
            all_text += "=" * 50
            all_text += "\n\n"

            all_text += text

        except Exception as e:

            print("Error:", file)
            print(e)


# 5. SAVE OCR TEXT


with open(
    ocr_output_file,
    "w",
    encoding="utf-8"
) as f:

    f.write(all_text)

print("\nOCR TEXT SAVED")


# 6. DIAGNOSIS EXTRACTION


diagnoses = []

possible_diagnoses = [
    "ACUTE GASTROENTERITIS WITH DEHYDRATION",
    "URINARY TRACT INFECTION",
    "DIABETES",
    "HYPERTENSION",
    "PNEUMONIA"
]

for diagnosis in possible_diagnoses:

    if diagnosis.lower() in all_text.lower():

        diagnoses.append(diagnosis)


# 7. PENDING RESULTS


pending_results = []

keywords = [
    "report awaited",
    "awaited",
    "pending",
    "culture report"
]

for keyword in keywords:

    if keyword.lower() in all_text.lower():

        pending_results.append(keyword)


# 8. CREATE SUMMARY


summary = ""

summary += "DISCHARGE SUMMARY\n"
summary += "=" * 50
summary += "\n\n"

summary += "DIAGNOSES:\n"

for d in diagnoses:

    summary += f"- {d}\n"

summary += "\n"

summary += "PENDING RESULTS:\n"

for p in pending_results:

    summary += f"- {p}\n"

# 9. SAVE SUMMARY

with open(
    summary_output_file,
    "w",
    encoding="utf-8"
) as f:

    f.write(summary)

print("\nSUMMARY SAVED")


# 10. DISPLAY RESULTS


print("\n")
print("=" * 50)
print("RESULTS")
print("=" * 50)

print("\nOCR FILE:")
print(ocr_output_file)

print("\nSUMMARY FILE:")
print(summary_output_file)

print("\nDIAGNOSES FOUND:")

for d in diagnoses:
    print("-", d)

print("\nPENDING RESULTS:")

for p in pending_results:
    print("-", p)

print("\nDONE")
```
